```python
#GET Stock price from yahoo finance
tsla = yf.Ticker("TSLA")
hist = tsla.history(period="7d", interval="1h")
```

```python
hist.index.name = 'date'
hist.reset_index(inplace=True)
```

```python
rows = []
for idx,row in hist.iterrows():
    date,open,close = row['date'],row['Open'],row['Close']
    p1 = influxdb_client.Point("tesla").tag("stock", "tesla").field('open', open)
    p2 = influxdb_client.Point("tesla").tag("stock", "tesla").field('close', close)
    p1.time(date)
    p2.time(date)
    write_api.write(bucket=bucket, org=org, record=p1)
    write_api.write(bucket=bucket, org=org, record=p2)
```
